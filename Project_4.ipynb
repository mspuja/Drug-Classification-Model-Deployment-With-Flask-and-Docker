# **Model Deployment with Flask**

```python
import flask
from flask import Flask, render_template, request
import pickle
import numpy as np
from pyngrok import ngrok
from flask_ngrok import run_with_ngrok
import warnings
warnings.filterwarnings('ignore')

app = Flask(__name__)
# Here, This is ones personal Authtoken. Which has to be exact ngrok config add-authtoken
ngrok.set_auth_token("2GGpWoygvnEkrmIETVxSElJIrt1_7XVpymwyGaZgiyEMeeaeT")  
run_with_ngrok(app)

model = pickle.load(open('model.pkl', 'rb'))

@app.route('/', methods=['GET'])
def home():
  return render_template('index.html')

@app.route('/', methods=['GET', "POST"])
def predict():
  input_values = [float(x) for x in request.form.values()]
  inp_features = [input_values]
  prediction = model.predict(inp_features)
  if prediction==1:
    return render_template('index.html', prediction_text='Predicted Drug Type is : DrugY')
  elif  prediction==2:
    return render_template('index.html', prediction_text='Predicted Drug Type is : drugX')
  elif  prediction==3:
    return render_template('index.html', prediction_text='Predicted Drug Type is : drugA')
  elif  prediction==4:
    return render_template('index.html', prediction_text='Predicted Drug Type is : drugC')
  elif  prediction==5:
    return render_template('index.html', prediction_text='Predicted Drug Type is : drugB')
  else:
    return render_template('index.html', prediction_text='Predicted Drug Type is : None')

app.run()
```
